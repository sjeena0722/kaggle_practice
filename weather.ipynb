```python
import pandas as pd
import os
```

```python
path = '파일 경로 입력'
file_list = os.listdir(path)
file_list_py = [file for file in file_list if file.endswith('.csv')] ## 파일명 끝이 .csv인 경우
```

```python
file_list_py 
```

```python
df = pd.DataFrame()
for i in file_list_py:
    data = pd.read_csv(path + i, encoding='cp949')
    df = pd.concat([df,data])
    
df = df.reset_index(drop = True)
df
```

```python
# int로 변환
df['지점'] = df['지점'].astype('int')

# 지점 -> 지역명
reg_dic = {540:'고양시', 423:'광명시', 569:'구리시',404:'김포시', 541:'남양주시', 572:'성남시', 532:'의정부시',  904:'부산광역시', 496:'세종특별자치시', 924:'울산광역시' , 864:'전라북도', 402:'하남시', 326:'양평군', 116:'안양시', 571:'화성시'  }
df['지점'] = df['지점'].map(reg_dic)

# 지점컬럼명 변경
df.rename(columns={'지점':'지역'}, inplace=True)

# 기온, 강수량 컬럼만 가져오기
df = df[['지역','일시', '기온(°C)', '강수량(mm)','습도(%)']]
```

```python
# csv파일 내보내기
df.to_csv('weather.csv')
```
